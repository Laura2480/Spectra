```python
def convert_video(file_name,processed_path):
  avi_path=join(processed_path,file_name)
  output_path=join(processed_path,file_name.split('.')[-2]+'.mp4')
  if exists(output_path):
    !rm -rf output_path
  !ffmpeg -y -loglevel info -i $avi_path $output_path
  !rm -rf $avi_path

def show_local_mp4_video(file_name, width=640, height=480):
  import io
  import base64
  from IPython.display import HTML
  video_encoded = base64.b64encode(io.open(file_name, 'rb').read())
  return HTML(data='''<video width="{0}" height="{1}" alt="test" controls>
                        <source src="data:video/mp4;base64,{2}" type="video/mp4" />
                      </video>'''.format(width, height, video_encoded.decode('ascii')))
```

```python
import imageio

# Percorso del file video
video_path = join(processed_path,file_list[2].split('.')[-2]+'.mp4')

# Indice del frame da estrarre
frame_index = 150

# Aprire il video
reader = imageio.get_reader(video_path)

# Leggere il frame specifico
frame = reader.get_data(frame_index)

# Salvare il frame come immagine
imageio.imwrite('extracted_frame.jpg', frame)

# Chiudere il lettore
reader.close()
```
